```python
df1['Current_Yr']=2022
df1.head()
```

```python
df1['Age']=df1['Current_Yr']-df1['Year']
df1.head()
```

```python
df2=df1.drop(['Year','Current_Yr','Present_Price','Seller_Type','Owner'],axis='columns')
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Fuel_Type=LabelEncoder()
le_Transmission=LabelEncoder()
```

```python
df2['Fuel_Type_N']=le_Fuel_Type.fit_transform(df2['Fuel_Type'])
df2['Transmission_N']=le_Transmission.fit_transform(df2['Transmission'])
df2.head()
```

```python
df3=df2.drop(['Fuel_Type','Transmission'],axis=1)
df3.head()
```

```python
df3.Car_Name=df3.Car_Name.apply(lambda x:x.strip())
```

```python
Counts=df3.Car_Name.value_counts(ascending=False)
Counts
```

```python
dummies=pd.get_dummies(df3['Car_Name'])
dummies.head()
```

```python
dummies1=dummies.drop(['xcent'],axis=1)
dummies1.head()
```

```python
df4=pd.concat([df3,dummies1],axis='columns')
df4.head()
```

```python
df5=df4.drop(['Car_Name'],axis=1)
df5.head()
```

```python
df5.shape
```

```python
X=df5.drop(['Selling_Price'],axis=1)
X.head()
```

```python
y=df5['Selling_Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X_test,y_test)
```

```python
model.score(X_test,y_test)
```

```python
def predict_price(Car_Name,Kms_Driven,Age,Fuel_Type_N,Transmission_N):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns)) 
    x[0] = Kms_Driven
    x[1] = Age
    x[2] = Fuel_Type_N
    x[3] = Transmission_N
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
