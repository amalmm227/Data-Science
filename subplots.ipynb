```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([8,3,9,2])

y= np.array([3,4,5,6])
plt.subplot(1,2,1)
plt.plot(x,y)
x=np.array([5,7,9])
y=np.array([1,3,6])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
x=np.array([11,22,33])
y=np.array([10,93,76])
plt.subplot(3,2,1)
plt.plot(x,y)
plt.title('a')


x=np.array([52,17,39])
y=np.array([13,53,67])
plt.subplot(3,2,2)
plt.plot(x,y)
plt.title('b')


x=np.array([57,17,99])
y=np.array([14,53,76])
plt.subplot(3,2,3)
plt.plot(x,y)
plt.title('c')



x=np.array([25,67,90])
y=np.array([51,63,66])
plt.subplot(3,2,4)
plt.plot(x,y)
plt.title('d')
plt.suptitle('figures')
plt.show()
```

```python
h=np.array([151,157,159,171,181,176,166,178,171])

w=np.array([67,76,99,109,56,78,88,77,99])

age = np.array([25,26,30,32,25,27,29,28,45])

plt.scatter(h,w,marker='>',color='g')
plt.xlabel('Height')
plt.ylabel('Weight')

plt.scatter(h,age)
plt.show()
```

```python
h=np.array([151,157,159,171,181,176,166,178,171])

w=np.array([67,76,99,109,56,78,88,77,99])

plt.bar(h,w,color='g',width=1.3)
plt.show()
#to show the graph horizontally use plt.barh
```

```python
d=np.random.normal(170,10,250)
print(d)
```

```python
plt.hist(d,bins=55,color='g')
plt.title('HISTOGRAM',color='r')
plt.xlabel('Height')
plt.ylabel('Age')


plt.show()
```

```python
#Pie chart
```

```python
q=np.array([20,25,30,35])
labels=['Apple','Banana','jackfruit','pineapple']
plt.pie(q,labels=labels)
plt.show
```

```python
q=np.array([20,25,30,35])
labels=['Apple','Banana','jackfruit','pineapple']
plt.pie(q,labels=labels,startangle=90)
plt.show
```

```python
q=np.array([20,25,30,35])
labels=['Apple','Banana','jackfruit','pineapple']
plt.pie(q,labels=labels,explode=[0.2,0.1,0.1,0.1])
plt.show
```

```python
q=np.array([20,25,30,35])
labels=['Apple','Banana','jackfruit','pineapple']
plt.pie(q,labels=labels,shadow=True)
plt.show
```

```python
q=np.array([20,25,30,35])
labels=['Apple','Banana','jackfruit','pineapple']
plt.pie(q,labels=labels,colors=['g','b','k','y'])
plt.legend(title='Labels',loc='upper left')
plt.show()
```
