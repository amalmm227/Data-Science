```python
#Calculate total sales data for last year for each product and show it in a pie chart
```

```python
df
```

```python
a1 = df['facecream'].sum()
a2 = df['facewash'].sum()
a3 = df['toothpaste'].sum()
a4 = df['bathingsoap'].sum()
a5 = df['shampoo'].sum()
a6 = df['moisturizer'].sum()
```

```python
label=['Facecream','Facewash','toothpaste','bathingsoap','shampoo','moisturizer']
plt.pie([a1,a2,a3,a4,a5,a6],labels=label,autopct='%.1f%%')
plt.legend(loc='upper left')
plt.show()
```
