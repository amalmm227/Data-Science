```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path=r'Sales_Data\\'
files=os.listdir(path)
```

```python
print(files)
```

```python
df1=pd.DataFrame()
for file in files:
    df1=pd.concat([df1,pd.read_csv(path+file)],ignore_index=True)
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1.isna().sum()
```

```python
df_na=df1[df1.isna().any(axis=1)]
df_na
```

```python
df2=df1.dropna()
df2.head()
```

```python
df2.shape
```

```python
df2['Month']=df2['Order Date'].str[0:2]
df2.head()
```

```python
df2[df2['Month']=='Or'].head()
```

```python
df3=df2[df2['Month']!='Or']
df3.head()
```

```python
df3[df3['Month']=='Or'].head()
```

```python
df3['Month']=df3['Month'].astype('int32')
df3.head()
```

```python
df4=df3.copy()
df4.head()
```

```python
df4['Quantity Ordered']=pd.to_numeric(df4['Quantity Ordered'])
df4['Price Each']=pd.to_numeric(df4['Price Each'])
df4.head()
```

```python
df4['Sales']=df4['Quantity Ordered']*df4['Price Each']
df4.head()
```

```python
result1=df4.groupby('Month').sum()
result1
```

```python
months=range(1,13)
plt.figure(figsize=(10,5))
plt.bar(months,result1['Sales'],color='r')
plt.xticks(months)
plt.xlabel('MONTHS')
plt.ylabel('SALES')
plt.show()
```

```python
df4.head(1)
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
df4['City']=df4['Purchase Address'].apply(lambda x: get_city(x)+' '+get_state(x))
df4.head(5)
```

```python
result2=df4.groupby('City').sum()
result2
```

```python
cities=[city for city,df in df4.groupby('City')]

plt.figure(figsize=(10,5))
plt.bar(cities,result2['Sales'])
plt.xticks(cities,rotation='vertical')
plt.xlabel('CITIES')
plt.ylabel('SALES IN US$')
plt.show()
```

```python
df4['Order Date']=pd.to_datetime(df4['Order Date'])
df4.head()
```

```python
df4['Hour']=df4['Order Date'].dt.hour
df4.head()
```

```python
df4.groupby('Hour').count()
```

```python
hours=[hour for hour,df in df4.groupby('Hour')]
plt.figure(figsize=(10,5))
plt.plot(hours,df4.groupby('Hour').count(),color='g')
plt.xticks(hours)
plt.xlabel('HOURS')
plt.ylabel(' NO OF SALES IN HOURS')
plt.grid()
```
