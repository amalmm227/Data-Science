```python
age= [12,9,11,10,5,9,4,7,1,2]
speed= [80,88,91,96,100,108,122,111,115,130]


from scipy import stats
m,c,r,p,std_err=stats.linregress(age,speed)

def f(age):
    
    line=m*age+c
    return line

p = list(map(f,age))


plt.scatter(age,speed,marker='+')
plt.plot(age,p)



plt.show() 
```
