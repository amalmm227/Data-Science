```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_fueltype=LabelEncoder()
le_carbody=LabelEncoder()
le_drivewheel=LabelEncoder()
le_enginetype=LabelEncoder()
```

```python
df7['F_type']=le_fueltype.fit_transform(df7['fueltype'])
df7['Carbody']=le_carbody.fit_transform(df7['carbody'])
df7['Drivewheel']=le_drivewheel.fit_transform(df7['drivewheel'])
df7['Engine_type']=le_enginetype.fit_transform(df7['enginetype'])
df7.head()
```

```python
df8=df7.drop(['fueltype','carbody','drivewheel','enginetype'],axis=1)
df8.head()
```

```python
dummies=pd.get_dummies(df8['CarName'])
dummies.head()
```

```python
dummies1=dummies.drop(['vw rabbit'],axis=1)
dummies1.head()
```

```python
df9=pd.concat([df8,dummies1],axis=1)
df9.head()
```

```python
df10=df9.drop(['CarName'],axis=1)
df10.head()
```

```python
X=df10.drop(['price'],axis=1)
X.head()
```

```python
y=df10['price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x_test,y_test)
```

```python
model.score(x_test,y_test)
```

```python
def predict_price(location,horsepower,Average_Mileage,F_type,Carbody,Drivewheel,Engine_type):    
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns)) 
    x[0] = horsepower
    x[1] = Average_Mileage
    x[2] = F_type
    x[3] = Carbody
    x[4] = Drivewheel
    x[5] = Engine_type
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
