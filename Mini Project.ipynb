```python
dfm.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
```

```python

dfm['le_Gender']=le.fit_transform(dfm['Gender'])

dfm['le_CustomerType']=le.fit_transform(dfm['CustomerType'])
dfm['le_TypeOfTravel']=le.fit_transform(dfm['TypeOfTravel'])
dfm['le_Class']=le.fit_transform(dfm['Class'])
dfm['le_satisfaction']=le.fit_transform(dfm['satisfaction'])
```

```python
dfm.head()
```

```python
dfm1 = dfm.drop(['Gender','CustomerType','TypeOfTravel','Class','satisfaction'],axis=1)
dfm1.head()
```

```python
# For male            = 0
# For customer Type   = loyal = 0, Disloyal=1
# Type of travel      = personal =1 ,business=0
# Class               = Eco plus =2 ,Business =0 ,Eco =1

#Satisfaction         = Neutral or Dissatisfied =0 ,Satisfied =1
```

```python
x = dfm1.drop(['le_satisfaction'],axis=1)
x.head()
```

```python
y = dfm1['le_satisfaction']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

# Decision Tree Implementation

```python
from sklearn.tree import DecisionTreeClassifier
model1 = DecisionTreeClassifier(criterion='entropy',random_state=0)
model1.fit(x_train,y_train)
model1.score(x_test,y_test)
```

# K- Neighbour Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(x_train,y_train)
model2.score(x_test,y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=100)
model3.fit(x_train,y_train)
model3.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model5 = GaussianNB()
model5.fit(x_train,y_train)
model5.score(x_test,y_test)
```

# SVM Classifier

```python
from sklearn.svm import SVC
model4 = SVC()
model4.fit(x_train,y_train)
model4.score(x_test,y_test)
```

# The best Score giving Algorithm for Male Customers


# Random Forest
# Score is = 95
