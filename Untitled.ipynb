```python
import pandas as pd
import os
```

```python
path=r'Sales_Data\\'
files=os.listdir(path)

print(files)
```

```python
df=pd.DataFrame()
for file in files:
    df=pd.concat([df,pd.read_csv(path+file)],ignore_index=True)
```

```python
df.head(9)
```

```python
df.shape
```

```python
df_na=df[df.isna().any(axis=1)]
df_na
```

```python
df1=df.dropna()
df1.head()
```

```python
df1.isna().sum()
```

```python
df1['Month']=df1['Order Date'].str[0:2]
df1.head(8)
```

```python
df1[df1['Month']=='Or'].head()
```

```python
df2=df1[df1['Month']!='Or']
df2.head()
```

```python
df2['Month']=df2['Month'].astype('int32')
df2.head()
```

```python
df2['Month'].describe
```

```python
df2.head()
```

```python
df2['Quantity Ordered']=pd.to_numeric(df2['Quantity Ordered'])
df2['Price Each']=pd.to_numeric(df2['Price Each'])
```

```python
df2.head()
```

```python
df2['Sales']=df2['Quantity Ordered']*df2['Price Each']
df2.head()
```

```python
result1=df2.groupby('Month').sum()
result1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months=range(1,13)
plt.figure(figsize=(10,5))
plt.bar(months,result1['Sales'],color='g')
plt.xticks(months)
plt.xlabel('MONTHS')
plt.ylabel('SALES IN US$')
plt.show()
```

```python
df2.head()
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]
```

```python
df2['City']=df2['Purchase Address'].apply(lambda x: get_city(x)+' '+get_state(x))
df2.head(11)
```

```python
result2=df2.groupby('City').sum()
result2
```

```python
cities=[city for city,df in df2.groupby('City')]

plt.figure(figsize=(10,5))
plt.bar(cities,result2['Sales'],color='g')
plt.xticks(cities,rotation='vertical')
plt.xlabel('CITIES')
plt.ylabel('SALES IN US$')
plt.show()
```
