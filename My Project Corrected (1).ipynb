```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_csv('car data.csv')
df1.head(20)
```

```python
df1.isna().sum()
```

```python
df1.info()
```

```python
df1['Present_Yr']=2022
df1.head()
```

```python
df1['Age']=df1['Present_Yr']-df1['Year']
df1.head()
```

```python
df2=df1.drop(['Year','Present_Price','Present_Yr'],axis='columns')
df2.head()
```

```python
df2['Selling_Price']=df2['Selling_Price']*100000
df2.head()
```

```python
df2['Selling_Price']=df2['Selling_Price'].astype('int32')
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Fuel_Type=LabelEncoder()
le_Seller_Type=LabelEncoder()
le_Transmission=LabelEncoder()
```

```python
df2['Fuel_Type_N']=le_Fuel_Type.fit_transform(df2['Fuel_Type'])
df2['Seller_Type_N']=le_Seller_Type.fit_transform(df2['Seller_Type'])
df2['Transmission_N']=le_Transmission.fit_transform(df2['Transmission'])
df2.head()
```

```python
df3=df2.drop(['Fuel_Type','Seller_Type','Transmission'],axis=1)
df3.head()
```

```python
df3['Car_Name']=df3['Car_Name'].apply(lambda x:x.strip())
```

```python
dummies=pd.get_dummies(df3['Car_Name'])
dummies.head()
```

```python
dummies1=dummies.drop(['xcent'],axis=1)
dummies1.head()
```

```python
df4=pd.concat([df3,dummies1],axis='columns')
df4.head()
```

```python
df5=df4.drop(['Car_Name'],axis=1)
df5.head()
```

```python
df5.shape
```

```python
X=df5.drop(['Selling_Price'],axis=1)
X.head()
```

```python
y=df5['Selling_Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
X.columns
```

```python
Car_Names='800'
np.where(X.columns==Car_Names)[0][0]
```

```python
def predict_price(Car_Name,Kms_Driven,Owner,Age,Fuel_Type_N,Seller_Type_N,Transmission_N):    
    loc_index = np.where(X.columns==Car_Names)[0][0]

    x = np.zeros(len(X.columns)) 
    x[0] = Car_Name
    x[1] = Kms_Driven
    x[2] = Owner
    x[3] = Age
    x[4] = Fuel_Type_N
    x[5] = Seller_Type_N
    x[6] = Transmission_N
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
